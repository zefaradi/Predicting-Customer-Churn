```r
telco <- read.csv("Telco-Customer-Churn.csv")
str(telco)
```

```r
telco$customerID <- as.factor(telco$customerID)
telco$gender <- as.factor(telco$gender)
telco$SeniorCitizen <- as.factor(telco$SeniorCitizen)
telco$Partner <- as.factor(telco$Partner)
telco$Dependents <- as.factor(telco$Dependents)
telco$PhoneService <- as.factor(telco$PhoneService)
telco$MultipleLines <- as.factor(telco$MultipleLines)
telco$InternetService <- as.factor(telco$InternetService)
telco$OnlineSecurity <- as.factor(telco$OnlineSecurity)
telco$OnlineBackup <- as.factor(telco$OnlineBackup)
telco$DeviceProtection <- as.factor(telco$DeviceProtection)
telco$TechSupport <- as.factor(telco$TechSupport)
telco$StreamingTV<- as.factor(telco$StreamingTV)
telco$StreamingMovies<- as.factor(telco$StreamingMovies)
telco$Contract<- as.factor(telco$Contract)
telco$PaperlessBilling<- as.factor(telco$PaperlessBilling)
telco$PaymentMethod <- as.factor(telco$PaymentMethod)
telco$Churn <- as.factor(telco$Churn)
```

```r
str(telco)
```

```r
sapply(telco, function(x) sum(is.na(x)))
```

```r
#Deleting the rows with NAs
telco <- telco[complete.cases(telco), ]
```

```r
##Data Processing
#Change “No internet service” to “No” for six columns, they are: “OnlineSecurity”, “OnlineBackup”, "Device Protection", "TechSupport","streamingTV","streamingMovies".
cols_recode1 <- c(10:15)
for(i in 1:ncol(telco[,cols_recode1])) {
telco[,cols_recode1][,i] <- as.factor(mapvalues
(telco[,cols_recode1][,i], from =c("No internet service"),to=c("No")))
}
```

```r
#Change “No phone service” to “No” for column “MultipleLines”
telco$MultipleLines <- as.factor(mapvalues(telco$MultipleLines,
from=c("No phone service"), to=c("No")))

min(telco$tenure)
max(telco$tenure)
```

```r
#The minimum tenure is 1 month and maximum tenure is 72 months, we can group them into 6 tenure groups: “0–1 year”, “1-2 years”, “2-3 years”, “3-4 years”, “4-5 years”, “5-6 years”. 

group_tenure <- function(tenure){
    if (tenure >= 0 & tenure <= 12){
        return('0-1 yr')
    }else if(tenure > 12 & tenure <= 24){
        return('1-2 yr')
    }else if (tenure > 24 & tenure <= 36){
        return('2-3 yr')
    }else if (tenure > 36 & tenure <=48){
        return('3-4 yr')
    }else if (tenure > 48 & tenure <=60){
        return('4-5 yr')
    }else if (tenure > 60 & tenure <=72){
        return('5-6 yr')
    }
}
```

```r
telco$tenure_group <- sapply(telco$tenure,group_tenure)
telco$tenure_group <- as.factor(telco$tenure_group)
```

```r
#Converting Senior Citizen variable into a cateogirical variable
telco$SeniorCitizen <- as.factor(ifelse(telco$SeniorCitizen==1, 'YES', 'NO'))
```

```r
#Remove the columns we do not need for the analysis
telco$customerID <- NULL
telco$tenure <- NULL
```

```r
#The Monthly Charges and Total Charges are correlated. So one of them will be removed from the model. We remove Total Charges.
telco$TotalCharges <- NULL
```

```r
#Visualizing the spread of tenure groups
options(repr.plot.width =6, repr.plot.height = 3)
ggplot(telco, aes(tenure_group, fill = tenure_group)) + geom_bar()+ theme(legend.position = "none")
```

```r
##All categorical variables have a reasonable broad distribution, therefore, all of them will be kept for the further analysis.
##LOGISTIC REGRESSION MODEL FITTING
# Split the data into training and testing sets
intrain<- createDataPartition(telco$Churn,p=0.7,list=FALSE)
set.seed(2018)
training<- telco[intrain,]
testing<- telco[-intrain,]
```

```r
#Confirm the splitting is correct
dim(training); dim(testing)

#Fitting the Model using all the variables
LogModel <- glm(Churn~.,family=binomial(link="logit"),data=training)
```

```r
print(summary(LogModel))
```

```r
#Using stepAIC variable selection to add or remove variables which gives the best performing model
LogModel2<- stepAIC(LogModel , direction="both")
summary(LogModel2)
```

```r
#Use variance inflation factor (vif) to get rid of redundant predictors or the 
#variables that have high multicollinearity between them

vif(LogModel2)
```

```r
#Removing DeviceProtection and OnlineSecurity due to high p-value

LogModel3<- glm(formula = Churn ~ SeniorCitizen + MultipleLines + InternetService + 
    TechSupport + StreamingTV + 
    StreamingMovies + Contract + PaperlessBilling + PaymentMethod + 
    MonthlyCharges + tenure_group, family = binomial(link = "logit"), 
    data = training)

summary(LogModel3)
vif(LogModel3)
```

```r
#LogModel3 all has significant variables, we'll use it for prediction first
#Model evaluation
## Assessing the predictive ability of the model
testing$Churn <- as.character(testing$Churn)
testing$Churn[testing$Churn=="No"] <- "0"
testing$Churn[testing$Churn=="Yes"] <- "1"
pred <- predict(LogModel3, type = "response", newdata = testing)
summary(pred)
testing$prob <- pred
```

```r
# Using probability cutoff of 50%.

pred_churn <- factor(ifelse(pred >= 0.50, "Yes", "No"))
actual_churn <- factor(ifelse(testing$Churn==1,"Yes","No"))
table(actual_churn,pred_churn)
```

```r
#Finding accuracy, sensitibity and specificity using 50% cutoff
cutoff_churn <- factor(ifelse(pred >=0.50, "Yes", "No"))
conf_final <- confusionMatrix(cutoff_churn, actual_churn, positive = "Yes")
accuracy <- conf_final$overall[1]
sensitivity <- conf_final$byClass[1]
specificity <- conf_final$byClass[2]
accuracy
sensitivity
specificity
```

```r
#Finding a different cutoff point to have a better sensitivity

perform_fn <- function(cutoff) 
{
  predicted_churn <- factor(ifelse(pred >= cutoff, "Yes", "No"))
  conf <- confusionMatrix(predicted_churn, actual_churn, positive = "Yes")
  accuray <- conf$overall[1]
  sensitivity <- conf$byClass[1]
  specificity <- conf$byClass[2]
  out <- t(as.matrix(c(sensitivity, specificity, accuray))) 
  colnames(out) <- c("sensitivity", "specificity", "accuracy")
  return(out)
}
```

```r
options(repr.plot.width =8, repr.plot.height =6)
summary(pred)
s = seq(0.01,0.80,length=100)
OUT = matrix(0,100,3)

for(i in 1:100)
{
  OUT[i,] = perform_fn(s[i])
} 

plot(s, OUT[,1],xlab="Cutoff",ylab="Value",cex.lab=1.5,cex.axis=1.5,ylim=c(0,1),
     type="l",lwd=2,axes=FALSE,col=2)
axis(1,seq(0,1,length=5),seq(0,1,length=5),cex.lab=1.5)
axis(2,seq(0,1,length=5),seq(0,1,length=5),cex.lab=1.5)
lines(s,OUT[,2],col="darkgreen",lwd=2)
lines(s,OUT[,3],col=4,lwd=2)
box()
legend("bottom",col=c(2,"darkgreen",4,"darkred"),text.font =3,inset = 0.02,
       box.lty=0,cex = 0.8, 
       lwd=c(2,2,2,2),c("Sensitivity","Specificity","Accuracy"))
abline(v = 0.305, col="red", lwd=1, lty=2)
axis(1, at = seq(0.1, 1, by = 0.1))
```

```r
#Re-predict the model with a cutoff value of 0.305

cutoff_churn <- factor(ifelse(pred >=0.305, "Yes", "No"))
conf_final <- confusionMatrix(cutoff_churn, actual_churn, positive = "Yes")
accuracy <- conf_final$overall[1]
sensitivity <- conf_final$byClass[1]
specificity <- conf_final$byClass[2]
accuracy
sensitivity
specificity
```

```r
#Logistic regression with a cutoff probability of 0.305 gives better
# sensitivity, accuracy and specificity

##DECISION TREE
Dtree <- ctree(Churn~., training)
pred_tree <- predict(Dtree, testing)
print("Confusion Matrix for Decision Tree"); table(Predicted = pred_tree, Actual = testing$Churn)
```
